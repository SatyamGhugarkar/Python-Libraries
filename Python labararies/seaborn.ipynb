Relational plot

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
# scatter plot is axis level
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

```python
# rel plot is figure level
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

```python
import plotly.express as px
p=px.data.gapminder()
p
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')#axes level
```

```python
temp_df=p[p['country'].isin(['India','Brazil','Germany'])]
temp_df

```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')#figure level
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

```python
# facet plot

sns.relplot(data=p,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

```python
# distrbution plot
# 1)distplot #figure level

tips=sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips,x='total_bill')
```
