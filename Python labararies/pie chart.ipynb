# Pie chart

Pie chart is a popular and visually intuitive tool used in data representation, making complex information easier to understand at a glance.
This circular graph divides data into slices, each representing a proportion of the whole, allowing for a clear comparison of different categories making it easier to digest complex information through a straightforward, intuitive format.

```python
import matplotlib.pyplot as plt
collection =[1000,2000,3000,70000,8000]
years= [2015,2016,2017,2018,2019]
plt.pie(collection,labels=years,autopct='%.1f%%')
```

```python
plt.style.use('classic')
collection =[1000,2000,3000,70000,8000]
years= [2015,2016,2017,2018,2019]
plt.pie(collection,labels=years,autopct='%0.5f%%',shadow=True,explode=[0,2,0,0,0])
```

```python
month=['jan','feb','march','april']
revenue=[100,150,120,170]
expences=[80,110,100,140]
plt.subplot(1,3,1)
plt.bar(month,revenue)

plt.subplot(1,3,2)
plt.pie(expences,labels=month)

plt.subplot(1,3,3)
plt.plot(month,revenue)
plt.tight_layout()
```

```python
days=['mon','tue','wed','thur']
temps=[30,32,35,33]
plt.plot(days,temps,marker='o')
plt.annotate('higher temp',xy=('wed',35),xytext=('thur',35),arrowprops=dict(facecolor='black'))
plt.savefig('sample.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'],color='blue')
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='orange')
```

```python
#second way to draw subplot
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(15,6),sharex=True)
ax[0].scatter(sample_df['avg'],sample_df['strike_rate'],color='red')
ax[0].set_title('Scatter Plot AVG/Strike rate')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(sample_df['avg'],sample_df['strike_rate'],color='red')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Strike Rate')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,6),sharex=True)
ax[0,0].scatter(sample_df['avg'],sample_df['strike_rate'])

ax[0,1].scatter(sample_df['runs'],sample_df['strike_rate'])

ax[1,0].hist(sample_df['avg'])

ax[1,1].hist(sample_df['runs'])
```
