# Box plot

```python

sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

# violin plo

```python

sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

# Bar plot

```python

sns.barplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
import numpy as np
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',estimator=np.std,errorbar=None)
```

```python

sns.pointplot(data=tips,x='day',y='total_bill',hue='sex')
```

# Count plot

```python

sns.countplot(data=tips,x='sex',hue='smoker')
```

# facet graph

```python

sns.catplot(data=tips,x='sex',y='total_bill',col='day',kind='bar',row='time')
```

# FacedGrid plot

```python
g=sns.FacetGrid(data=tips,col='day',row='time',hue='sex')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris,hue='species')
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
