```python
min=float('inf')
max=float('-inf')
for i in range(len(aa)):
    for j in range(len(aa[0])):
        if min>aa[i][j]:
            min=aa[i][j]
        if max<aa[i][j]: 
            max=aa[i][j]

print(min)
print(max)
```
