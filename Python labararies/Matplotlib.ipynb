```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c","c++","java"]
y = [85,70,60,82]
z=[56,34,40,60]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("wscube",fontsize=15)

plt.bar(p,y,width,color="y",align="center",label="wscube")#,linewidth=3,linestyle=":",alpha=0.5)
plt.bar(p1,z,width,color="g",align="center",label="ws")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

#3D surface plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
print(xx)

z=xx**2+yy**2
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=np.sin(xx)+np.cos(yy)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
#  log
import numpy as np
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=np.sin(xx)+np.log(yy)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
# contour
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
print(xx)

z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot()
ax.contour(xx,yy,z,cmap='viridis')
```
